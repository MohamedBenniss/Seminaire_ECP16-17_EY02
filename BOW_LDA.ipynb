# D√©compsition en Valeur Singuli√®re

la decomposition en valeurs singuli√®re: M=USV, o√π
U:  est une matrice reelle orthogonale, (appartenant √† l'√©space des commentaires)
S: matrice rectangulaire diagonale, avec les valeurs singuli√®res en diagonale ordonn√©e du plus grande au plus petite
V: real or complex unitary matrix, (appartenant √† l'√©space des )

la decomposition effectu√©e ici effectue un seuillage sur les valeurs singuli√®res et ne garde que les k plus grandes valeurs singuli√®re dans S (ainsi que les colonnes ou les lignes correspndantes aux valeurs gard√©es dans U et V) 

```python
from scipy.sparse import linalg
K=10
u, s0, v = linalg.svds(tfidf, k = K)

s0

u

SV=np.zeros((K,K))
np.fill_diagonal(SV,s0)

SV

VT=np.dot(np.transpose(v),SV)

VT.shape
```

```python
# Tokenize the documents.

from nltk.tokenize import RegexpTokenizer

# Split the documents into tokens.
tokenizer = RegexpTokenizer(r'\w+')

for i in range(len(tweetText)):
    tweetText[i] = tweetText[i].lower()  # Convert to lowercase.
    tweetText[i] = tokenizer.tokenize(tweetText[i])  # Split into words.


# Remove numbers, but not words that contain numbers.
tweetText = [[token for token in tweet if not token.isnumeric()] for tweet in tweetText]

# Remove words that are only one character.
tweetText = [[token for token in tweet if len(token) > 1] for tweet in tweetText]
```

```python
print(tweetText)
len(tweetText)
```

```python
# Lemmatize the documents.

from pattern.fr import lemma

# Lemmatize all words in tweets
tweetText = [[lemma(token) for token in tweet] for tweet in tweetText]
```

```python
# Compute bigrams.
# in order to keep all significant association of two words
from gensim.models import Phrases

# Add bigrams and trigrams to docs (only ones that appear nBG times or more).
nBG=10
bigram = Phrases(tweetText, min_count=nBG)
for i in range(len(tweetText)):
    for token in bigram[tweetText[i]]:
        if '_' in token:
            # Token is a bigram, add to document.
            tweetText[i].append(token)
```

```python
# Remove rare and common tokens.

from gensim.corpora import Dictionary

# Create a dictionary representation of the documents.
dictionary = Dictionary(tweetText)

# Filter out words that occur less than Nmin tweets, or more than Maxp of the documents.
Nmin=7
Maxp=0.6
dictionary.filter_extremes(no_below=Nmin, no_above=Maxp)
```

```python
# Vectorize data.

# Bag-of-words representation of the documents.
corpus = [dictionary.doc2bow(tweet) for tweet in tweetText]
```

```python
print('Number of unique tokens: %d' % len(dictionary))
print('Number of documents: %d' % len(corpus))
```

```python
# Train LDA model.

from gensim.models import LdaModel

# Set training parameters.
num_topics = 10
chunksize = 2000
passes = 20
iterations = 400
eval_every = None  # Don't evaluate model perplexity, takes too much time.

# Make a index to word dictionary.
temp = dictionary[0]  # This is only to "load" the dictionary.
id2word = dictionary.id2token

%time model = LdaModel(corpus=corpus, id2word=id2word, chunksize=chunksize, \
                       alpha='auto', eta='auto', \
                       iterations=iterations, num_topics=num_topics, \
                       passes=passes, eval_every=eval_every)
```

```python
top_topics = model.top_topics(corpus, num_words=20)

# Average topic coherence is the sum of topic coherences of all topics, divided by the number of topics.
avg_topic_coherence = sum([t[1] for t in top_topics]) / num_topics
print('Average topic coherence: %.4f.' % avg_topic_coherence)

from pprint import pprint
pprint(top_topics)
```
